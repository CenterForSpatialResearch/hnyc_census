```python
def seperate_column(df, column):
    df['direction'] = df[column].apply(lambda x: ' '.join(re.findall(pattern['direction'], x)))
    df['end'] = df[column].apply(lambda x: ' '.join(re.findall(pattern['end'], x)))
    df['street_type'] = df[column].apply(lambda x: ' '.join(re.findall(pattern['street_type'], x)))
    df['street_number'] = df[column].str.extract('(\d+)')
    df['street_word_number'] = df[column].apply(lambda x: ' '.join(re.findall(pattern['word_number'], x)))
    df['street_name'] = df[column].apply(lambda x: ' '.join(re.findall(pattern['street_name'], x)))
    
    return df
    
    
    
```

```python
seperate_column(df, 'Street address 2')
```
