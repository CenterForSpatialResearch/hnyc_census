```python
#import the packages that needed
import pandas as pd
import numpy as np
import re
import warnings
warnings.filterwarnings('ignore')
```

```python
#Method 1: Regular expression
def clean(x):
    x = str(x.upper())
    pattern = re.compile(r'STREET|SRT$|SR$|SRT$|SST$|SEET$|TREET$|SHEER$|SHEE$|STREE$|SREET$|REET$|STEE$|ST$')
    x = re.sub(pattern, "ST", x)
    pattern = re.compile(r'^WEST|^W|^WST')
    x = re.sub(pattern, "W", x)
    pattern = re.compile(r'^N|^NORTH')
    x = re.sub(pattern, "N", x)
    pattern = re.compile(r'^S|^SOUTH')
    x = re.sub(pattern, "S", x)
    pattern = re.compile(r'^E|^EAST|^EST')
    x = re.sub(pattern, "E", x)
    pattern = re.compile(r'DRIVE|DR|DV|DE$|DRV|DRI|DRIV|DRIE')
    x = re.sub(pattern, "DR", x)
    pattern = re.compile(r'CIRCLE|CIR|CRL|CIRC|CR|CL|CIRCL|CICLE')
    x = re.sub(pattern, "CIR", x)
    pattern = re.compile(r'AVENUE|AVE|AV|AVN|AVEN|AVENU')
    x = re.sub(pattern, "AVE", x)
    pattern = re.compile(r"(?<=\\s)COURT|CT|CRT|CRT|CTR|(?<=\\s)COUR|(?<=\\s)COT|(?<=\\s)CORT")
    x = re.sub(pattern, "CT", x)
    pattern = re.compile(r"BOULEVARD|BLVD|BVLD|BV|BLD|BD|BL|BLV")
    x = re.sub(pattern, "BLVD", x)
    pattern = re.compile(r"ROAD|RD|RAD|ROD")
    x = re.sub(pattern, "RD", x)
    pattern = re.compile(r"ALLEY|ALY|AL|ALLY|ALEY|ALLE|AY")
    x = re.sub(pattern, "BLVD", x)
    pattern = re.compile(r"PLACE|PL|PLC|PLE|PC|PLAC|PLCE|PCE")
    x = re.sub(pattern, "PL", x)
    pattern = re.compile(r"PK|PRK|PRAK|PAK")
    x = re.sub(pattern, "PARK", x)
    pattern = re.compile(r"PARKWAY|PKWY|PARKW|PWY|PKW|PRKWY|PKWY|PKW")
    x = re.sub(pattern, "PKWY", x)
    pattern = re.compile(r"APPR|APR|APPROA|APRCH|APPRCH")
    x = re.sub(pattern, "APPROACH", x)
    pattern = re.compile(r"TERRACE|TER|TERR|TRC|TRCE|TR")
    x = re.sub(pattern, "TER", x)
    pattern = re.compile(r"PLAZA|PLZ|PLAZ|PZ|PLZA")
    x = re.sub(pattern, "PLZ", x)
    pattern = re.compile(r"LANE|LN|LNE|LAN")
    x = re.sub(pattern, "LN", x)
    #pattern = re.compile("\\<BRIDGE\\>|\\<BRG\\>|\\<BRGD\\>|\\<BGE\\>","BRG",x)
    pattern = re.compile(r"HILL|HL|HLL|HIL")
    x = re.sub(pattern, "HL", x)
    pattern = re.compile(r"HEIGHTS|HTS|HT|HEIGHT|HEGHTS|HHT|HEIGT") 
    x = re.sub(pattern, "HTS", x)
    pattern = re.compile(r"SLP|SLEP|SLIIP|SLI")
    x = re.sub(pattern, "SLIP", x)
    pattern = re.compile(r"RON|RW|ROE|ROOW")
    x = re.sub(pattern, "ROW", x)
    pattern = re.compile(r"SQUARE") 
    x = re.sub(pattern, "SQ", x)
    pattern = re.compile(r".*\\((.*)\\).*")
    x = re.sub(pattern, "1", x)
    pattern = re.compile(r"(\s)(ST)") #remove ST after space
    x = re.sub(pattern, " ", x)
    pattern = re.compile(r"\\d+\\ - *\\d*|\\d+\\ TO *\\d*|\\d+\\-\\d*") #remove addresses
    x = re.sub(pattern, " ", x)
    
    # deal with numbered streets
    pattern = re.compile(r'[\d](ST|ND|RD|TH)\b')
    x = re.sub(pattern, "1", x)
    x.replace("(?=[0-9])(ST|ND|RD|TH)", " ")
    pattern = re.compile(r'ONE HUNDRED|ONEHUNDRED|HUNDRED|HUDRED|HUNDED')
    x = re.sub(pattern, "1", x)
    pattern = re.compile(r'TWO HUNDRED|TWOHUNDRED')
    x = re.sub(pattern, "2", x)
#     pattern = re.compile(r'-|\\|\'|\.|\*|\d')
#     x = re.sub(pattern, " ", x)
    pattern = re.compile(r'&')
    x = re.sub(pattern, "AND", x)
    pattern = re.compile(r'1ST|b1')
    x = re.sub(pattern, "1", x)
    pattern = re.compile(r'2ND|b2')
    x = re.sub(pattern, "2", x)
    pattern = re.compile(r'3RD|b3')
    x = re.sub(pattern, "3", x)
    pattern = re.compile(r'4TH|b4')
    x = re.sub(pattern, "4", x)
    pattern = re.compile(r'5TH|b5')
    x = re.sub(pattern, "5", x)
    pattern = re.compile(r'6TH|b6')
    x = re.sub(pattern, "6", x)
    pattern = re.compile(r'7TH|b7')
    x = re.sub(pattern, "7", x)
    pattern = re.compile(r'8TH|b8')
    x = re.sub(pattern, "8", x)
    pattern = re.compile(r'9TH|b9')
    x = re.sub(pattern, "9", x)
    pattern = re.compile(r'TENTH|10TH|b10')
    x = re.sub(pattern, "10", x)
    pattern = re.compile(r'ELEVENTH|11TH|b11')
    x = re.sub(pattern, "11", x)
    pattern = re.compile(r'TWELFTH|12TH|b12')
    x = re.sub(pattern, "12", x)
    pattern = re.compile(r'THIRTEENTH|13TH|b13')
    x = re.sub(pattern, "13", x)    
    pattern = re.compile(r'FORTEENTH|14TH|b14')
    x = re.sub(pattern, "14", x)
    pattern = re.compile(r'FIFTEENTH|15TH|b15')
    x = re.sub(pattern, "15", x)
    pattern = re.compile(r'SIXTEENTH|16TH|b16')
    x = re.sub(pattern, "16", x)
    pattern = re.compile(r'SEVENTEENTH|17TH|b17')
    x = re.sub(pattern, "17", x)
    pattern = re.compile(r'EIGHTEENTH|18TH|b18')
    x = re.sub(pattern, "18", x)
    pattern = re.compile(r'NINETEENTH|19TH|b19')
    x = re.sub(pattern, "19", x)
    pattern = re.compile(r'TWENTY|TWENTI|TENTI')
    x = re.sub(pattern, "2", x)
    pattern = re.compile(r'THIRTY|THIRTHY|THIRTEY|TIRTY|TRITHY')
    x = re.sub(pattern, "3", x)
    pattern = re.compile(r'FORTY|FOURTY|FOURTHY|FRTY')
    x = re.sub(pattern, "4", x)    
    pattern = re.compile(r'FIFTY|FIFTEY|FIFT|FITY|FIFTHY')
    x = re.sub(pattern, "5", x) 
    pattern = re.compile(r'SIXTY|SXTY|SIXY|SXTY|SIXTHY|SIXTEY')
    x = re.sub(pattern, "6", x)  
    pattern = re.compile(r'SEVENT|SEVENTY|SEVENTEY|SVENTY|SEVENTI')
    x = re.sub(pattern, "7", x)  
    pattern = re.compile(r'EIGHTY|EIGHTEY|EIGHTE')
    x = re.sub(pattern, "8", x)  
    pattern = re.compile(r'UNITY|NINTH|NINETY|NINETEY|NINETIETH|NINTY')
    x = re.sub(pattern, "9", x)  
    pattern = re.compile(r'TWENTIETH|TWENTIEFTH')
    x = re.sub(pattern, "20", x)
    pattern = re.compile(r'THIRTIETH|THIRTIEFTH')
    x = re.sub(pattern, "30", x)
    pattern = re.compile(r'FORTIETH|FOURTIETH|FOURTHENTH')
    x = re.sub(pattern, "40", x)
    pattern = re.compile(r'FIFTIETH')
    x = re.sub(pattern, "50", x)
    pattern = re.compile(r'SIXTIETH')
    x = re.sub(pattern, "60", x)
    pattern = re.compile(r'SEVENTIETH')
    x = re.sub(pattern, "70", x)
    pattern = re.compile(r'EIGHTIETH|EIGHTETH')
    x = re.sub(pattern, "80", x)
    pattern = re.compile(r'NINETIETH|NINTIETH')
    x = re.sub(pattern, "90", x)
    pattern = re.compile(r'^\bST')
    x = re.sub(pattern, "SAINT", x)
    pattern = re.compile(r'\bHOUSE\b|\bHOSTEL\b|\bHOTEL\b|\bLODGE\b|\bLODGING\b')
    x = re.sub(pattern,"", x)
    x = x.strip()
    
    # overall spacing problem extra white space
#     pattern = re.compile(r'((\\w+|\\d+)\\s\\s(\\w+|\\d+))')
#     x = re.sub(pattern, '\\2 \\3', x)
#     pattern = re.compile(r'((\\w+|\\d+)\\s\\s\\s(\\w+|\\d+))')
#     x = re.sub(pattern, '\\2 \\3', x)
    
    # 4 W to W 4
#     pattern = re.compile(r'(\\w+|\\d+)\\s(\\bN\\b|\\bW\\b|\\bS\\b|\\bE\\b)')
#     x = re.sub(pattern, '\\3\\2 \\1', x)
    
    # AVE after A-Z and Digits
#     pattern = re.compile(r'(AVE)|(b[A-Z])')
#     x = re.sub(pattern, '\\3\\2 \\1',x)
#     pattern = re.compile(r'(AVE)\\s(\\b\\d+)')
#     x = re.sub(pattern, '\\3\\2 \\1', x)
    
    ###### ALL MANUAL CLEANING GOES BELOW HERE ######
    
    # Manual Cleaning
    pattern = re.compile(r'ALLANTICï½œATLASTA')
    x = re.sub(pattern, "ATLANTIC", x)
    pattern = re.compile(r'ALLEM')
    x = re.sub(pattern, "ALLEN", x)    
    pattern = re.compile(r'ANENUE')
    x = re.sub(pattern, "AVE", x)    
    pattern = re.compile(r'CROTON|GROTON')
    x = re.sub(pattern, "AUDUBON", x)    
    pattern = re.compile(r'AT RINS')
    x = re.sub(pattern, "ATKINS", x)     
    
    pattern = re.compile(r'BATTIE')
    x = re.sub(pattern, "BALTIC", x)      
    pattern = re.compile(r'BARREE')
    x = re.sub(pattern, "BARROW", x)   
    pattern = re.compile(r'BESSHLEY')
    x = re.sub(pattern, "BURLING", x)       
    pattern = re.compile(r'BIRY|B WAY|BWAY|BRAKSWAY')
    x = re.sub(pattern, "BROADWAY", x)         
    pattern = re.compile(r'BUSTWICK')
    x = re.sub(pattern, "BUSHWICK", x) 
    pattern = re.compile(r'BUTTER')
    x = re.sub(pattern, "BUTLER", x) 
    pattern = re.compile(r'BREEVORT')
    x = re.sub(pattern, "BREVOORT", x) 
    pattern = re.compile(r'BRENNEL|BROOMES|BROOM|BRANNAS|BROWN')
    x = re.sub(pattern, "BROOME", x) 
    pattern = re.compile(r'BLACKER|BLENKER')
    x = re.sub(pattern, "BLEECKER", x) 
    
    pattern = re.compile(r'CLAIR')
    x = re.sub(pattern, "CLASSON", x) 
    pattern = re.compile(r'CLISTEN')
    x = re.sub(pattern, "CLINTON", x)     
    pattern = re.compile(r'CHERY')
    x = re.sub(pattern, "CHERRY", x) 
    pattern = re.compile(r'CHRYSTEE|CHRYSTAL|CHTYSTIE|CHRYSTEL')
    x = re.sub(pattern, "CHRYSTIE", x)     
    pattern = re.compile(r'CARRAL|COYAL')
    x = re.sub(pattern, "CANAL", x)    
    pattern = re.compile(r'COLUMBIN')
    x = re.sub(pattern, "COLUMBIA", x) 
    pattern = re.compile(r'CAMNON')
    x = re.sub(pattern, "CANNON", x) 
    pattern = re.compile(r'CROWH')
    x = re.sub(pattern, "CROWN", x) 
    
    pattern = re.compile('DEVAL')
    x = re.sub(pattern, "DEVOE", x)
    pattern = re.compile('DEBOUCHEL')
    x = re.sub(pattern, "DEBEVOISE",x)
    pattern = re.compile('DAFONT')
    x = re.sub(pattern, "DUPONT", x)
    pattern = re.compile('DEGRAN')
    x = re.sub(pattern, "DEGRAW", x)
    pattern = re.compile('DENBO|DEKALB')
    x = re.sub(pattern, "DE KALB", x)
    pattern = re.compile('DELAMERE|DALANEY|DELANEY|DELANCY')
    x = re.sub(pattern, "DELANCEY", x) 
    
    pattern = re.compile('ELTHZROTH|ELLSWICK')
    x = re.sub(pattern, "ELLIOTT", x)
    pattern = re.compile('ELDREDGE|CLARIDGE')
    x = re.sub(pattern, "ELDRIDGE", x) 
    pattern = re.compile('ESSEY')
    x = re.sub(pattern, "ESSEX", x) 
    pattern = re.compile('FORSYTHE')
    x = re.sub(pattern, "FORSYTH", x) 
    pattern = re.compile('FLATHISH')
    x = re.sub(pattern, "FLATBUSH", x)

    pattern = re.compile('GLANCE')
    x = re.sub(pattern, "GRAND", x) 
    pattern = re.compile('GOAST')
    x = re.sub(pattern, "GOERCK", x)
    pattern = re.compile('GREENS')
    x = re.sub(pattern, "GREENE", x)
    pattern = re.compile('GREENRICH|GAMWICH')
    x = re.sub(pattern, "GREENWICH", x) 

    pattern = re.compile('HOUTON')
    x = re.sub(pattern,  "HOUSTON", x) 
    pattern = re.compile('HAVES')
    x = re.sub(pattern, "HEWES", x) 
    pattern = re.compile('HAKEY')
    x = re.sub(pattern, "HALSEY", x)
    pattern = re.compile('HEWEY')
    x = re.sub(pattern, "HENRY", x)
    pattern = re.compile('HICK')
    x = re.sub(pattern, "HICKS", x)
    pattern = re.compile('HUMBOLOT|HUMBARD|HUMBOLT')
    x = re.sub(pattern, "HUMBOLDT", x) 
    
    pattern = re.compile('JOHOM')
    x = re.sub(pattern, "JOHNS", x) 
    
    pattern = re.compile('KIOP|HEAP')
    x = re.sub(pattern, "KEAP", x) 

    pattern = re.compile('LAY FAY ESTE|LADORATT|LAFYAYETTE')
    x = re.sub(pattern, "LAFAYETTE", x)
    pattern = re.compile('LIRA|LOUMOR|LARMER')
    x = re.sub(pattern, "LORIMER", x)
    pattern = re.compile('LAAVIUK')
    x = re.sub(pattern, "LAWRENCE", x) 
    pattern = re.compile('LAIDLOW')
    x = re.sub(pattern, "LUDLOW", x) 
    pattern = re.compile('TEX')
    x = re.sub(pattern, "LEX", x)
    pattern = re.compile('REPPERTS')
    x = re.sub(pattern, "LEFFERTS", x)

    pattern = re.compile('PARLE|MALLE|MYETTE')
    x = re.sub(pattern, "MYRTLE", x)
    pattern = re.compile('MC DOUGALL|MC DOUGAL|MCDOUGALL')
    x = re.sub(pattern, "MCDOUGAL", x)
    pattern = re.compile('MC DONOUGH')
    x = re.sub(pattern, "MCDONOUGH", x)
    pattern = re.compile('MANZA|MAREY')
    x = re.sub(pattern, "MARCY", x)
    pattern = re.compile('MESCOLE')
    x = re.sub(pattern, "MESEROLE", x)
    pattern = re.compile('MEASE')
    x = re.sub(pattern, "MOORE", x)
    pattern = re.compile('MEDDLER')
    x = re.sub(pattern, "MIDDLETON", x)
    pattern = re.compile('MANGEN')
    x = re.sub(pattern, "MANGIN", x)
    pattern = re.compile('HAULL|MALLERY')
    x = re.sub(pattern, "MULBERRY", x)

    pattern = re.compile('NAPOLK')
    x = re.sub(pattern, "NORFOLK", x)
    pattern = re.compile('VAST AND')
    x = re.sub(pattern, "NOSTRAND", x)

    pattern = re.compile('DAK')
    x = re.sub(pattern, "OAK", x)
    pattern = re.compile('OLWEN')
    x = re.sub(pattern, "OLIVER", x)
    pattern = re.compile('GERHARD')
    x = re.sub(pattern, "ORCHARD", x)

    pattern = re.compile('PUTT')
    x = re.sub(pattern, "PITT", x)
    pattern = re.compile('PERROTT|PERROTT PREMPONT')
    x = re.sub(pattern, "PIERREPONT", x)
    pattern = re.compile('PLAD')
    x = re.sub(pattern, "PLACE", x)
    pattern = re.compile('PRUFER')
    x = re.sub(pattern, "PROSPECT", x)
    pattern = re.compile('PREDIDUNT')
    x = re.sub(pattern, "PRESIDENT", x)
    pattern = re.compile('PALOKA')
    x = re.sub(pattern, "PULASKI", x)

    pattern = re.compile('RUTHIE')
    x = re.sub(pattern, "RUTLEDGE", x)
    pattern = re.compile('RIDAL')
    x = re.sub(pattern, "RIDGE", x)
    pattern = re.compile('RAYSON')
    x = re.sub(pattern, "RYERSON", x)
    pattern = re.compile('REVENTON')
    x = re.sub(pattern, "RIVINGTON", x)
    pattern = re.compile('RUALMAINE|RICER SIDE')
    x = re.sub(pattern, "RIVERSIDE", x)
    pattern = re.compile('REDERICK|RENNICK')
    x = re.sub(pattern, "RENWICK", x)

    pattern = re.compile('SELLTOWN')
    x = re.sub(pattern, "SULLIVAN", x)
    pattern = re.compile('SISH')
    x = re.sub(pattern, "SIDE", x)
    pattern = re.compile('STUCKER')
    x = re.sub(pattern, "STEUBEN", x)
    pattern = re.compile('STATES')
    x = re.sub(pattern, "STATE", x)
    pattern = re.compile('SCHAALS')
    x = re.sub(pattern, "SCHOLES", x)
    pattern = re.compile('SUMME')
    x = re.sub(pattern, "SUMMIT", x)
    pattern = re.compile('SCHOMERDOSA')
    x = re.sub(pattern, "SCHERMERHORN", x)
    pattern = re.compile('DOUTH|SONSE')
    x = re.sub(pattern, "SOUTH", x)
    pattern = re.compile('STUYVESTANT')
    x = re.sub(pattern, "STUYVESANT", x)

    pattern = re.compile('STONPSON')
    x = re.sub(pattern, "THOMPSON", x)
    pattern = re.compile('TRAY')
    x = re.sub(pattern, "TROY", x)
    pattern = re.compile('TAYLER')
    x = re.sub(pattern, "TAYLOR", x)

    pattern = re.compile('WMON')
    x = re.sub(pattern, "UNION", x)

    pattern = re.compile('WAR CAREN')
    x = re.sub(pattern, "VAN BUREN", x)
    pattern = re.compile('VEMON')
    x = re.sub(pattern, "VERNON", x)
    pattern = re.compile('VANDERLY|VANDERSLIDE')
    x = re.sub(pattern, "VANDERBILT", x)

    pattern = re.compile('WYONIA')
    x = re.sub(pattern, "WYONA", x)
    pattern = re.compile('WITKINS')
    x = re.sub(pattern, "WATKINS", x)
    pattern = re.compile('WALLWORTH')
    x = re.sub(pattern, "WALWORTH", x)
    pattern = re.compile('WHIPPER')
    x = re.sub(pattern, "WHIPPLE", x)
    pattern = re.compile('WALLABANK|WALKABOUT')
    x = re.sub(pattern, "WALLABOUT", x)
    pattern = re.compile('WASH|WASTEWATER')
    x = re.sub(pattern, "WASHINGTON", x) 

    return x
      
```
